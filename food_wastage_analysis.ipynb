```python
providers_clean = providers.copy()
receivers_clean = receivers.copy()
food_clean = food.copy()
claims_clean = claims.copy()
```

```python
providers_clean = providers_clean.drop_duplicates()
receivers_clean = receivers_clean.drop_duplicates()
food_clean = food_clean.drop_duplicates()
claims_clean = claims_clean.drop_duplicates()
```

```python
print(providers_clean.duplicated().sum())
print(receivers_clean.duplicated().sum())
print(food_clean.duplicated().sum())
print(claims_clean.duplicated().sum())
```

```python
providers_clean.isnull().sum()
receivers_clean.isnull().sum()
food_clean.isnull().sum()
claims_clean.isnull().sum()
```

```python
providers_clean = providers_clean.fillna("Unknown")
receivers_clean = receivers_clean.fillna("Unknown")
```

```python
food_clean["Quantity"] = food_clean["Quantity"].fillna(0)
```

```python
providers_clean.columns
```

```python
providers_clean.columns = providers_clean.columns.str.strip()
receivers_clean.columns = receivers_clean.columns.str.strip()
food_clean.columns = food_clean.columns.str.strip()
claims_clean.columns = claims_clean.columns.str.strip()
```

```python
providers_clean.dtypes
food_clean.dtypes
claims_clean.dtypes
```

```python
food_clean["Expiry_Date"] = pd.to_datetime(
    food_clean["Expiry_Date"]
)

```

```python
claims_clean["Timestamp"] = pd.to_datetime(
    claims_clean["Timestamp"]
)
```

```python
claims_clean["Claim_Date"] = pd.to_datetime(
    claims_clean["Claim_Date"]
)
```
